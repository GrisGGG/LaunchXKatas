# Entornos virtuales

Es un entorno aislado que permite a determinada aplicación ocupar las librerias y modulos instalados en este sin afectar las librerias ya instaladas en el sistema.

1.- Creamos la carpeta donde pondremos nuestro entorno virtual.

**mkdir kata02**

<img src="./img/creacion_carpeta.png">

2.- Ahora crearemos el entorno virtual mediante venv, con el siguiente comando.

**python3 -m venv env**

Con esto se nos creara una carpeta llamada env con las siguientes carpetas.

<img src="./img/carpetas.png">

3.- Se recomieda poner nuestros archivos del programa en una carpeta llamada src

4.- Activar nuestro entorno virtual, con el siguiente comando.

**env\Scripts\activate**

<img src="./img/creacion_entornoV.png">

5.- Sabremos que estamos dentro del entorno virtual cuando nos aparezca el nombre de nuestra carpeta antes de la dirección donde nos encontramos, ejemplo.

<img src="./img/entornoV.png">

6.- Con este comando podemos checar que librerias tenemos instaladas dentro del entorno virtual.

**pip freeze**

<img src="./img/pip.png">

7.- Para instalar librerias 

**pip install python-dateutil**

<img src="./img/instalar.png">


8.- Para salir del entorno virtual.
 
 **deactivate**

<img src="./img/desactivar.png">
